## 1 – Business Understanding

Predicting churn rates is a challenging and common problem that data scientists and analysts regularly encounter in any customer-facing business.
However, it is one that is worth being solved since it allows businesses to engage with users who are likely to churn before they actually do so and provide incentives to stay with the business. 
Knowing if a user might churn soon gives marketers the opportunity to introduce measures to avoid a particular's user churn. Possible actions include giving away free trial periods, discounts, or just further information about the business’ products.
<br>
Therefore, the goal of this project is to get a grasp of the data that is available for analysis and finding a way to predict customer shurn. To facilitate the process, three questions that are answered in this notebook are: 
1.  How often do users access which pages and when? > get an understanding of the high-level customer behaviour and service usage.
2. Which features are correlated with user churn? > find indications of users that are likely to churn in the future; those can then be adressed specifically. 
3. Can we predict churn from those features? > Find out if a fit model can generate usable predictions from the input data.

These questions are answered over the next 4 sections.

#### Define Churn

Churn is here defined as a churned user who successfully accessed the `Cancellation Confirmation` page; add column `churned` using `Cancellation Confirmation` page visit.

## 3 – Prepare Data
In the 3rd step of the CRISP-DM, the data is prepared in order to be in the right format for training a machine learning mode. 
For this purpose, the data will first be aggregated per `userId`, then outliers (defined here as smaller than the 5th quantile and larger than the 95th quantile) will be set the bounds (5th/95th quantile), columns with skewed data are log scaled, and all columns are standardized. <br>
Finally, all feature columns are combined into one column. 

### Aggregate dataframe to user level

```python
aggLogDataByUser = (logDataChurnInfo
                    .withColumn("latestLevel", F.last("level").over(Window.partitionBy("userId").orderBy("datetime")))
                    .filter(F.col("page").isin("NextSong", "Roll Advert", "Thumbs Down", "Add Friend", 
                                               "Settings", "Help", "Add to Playlist"))
                    .groupBy("userId") # Aggregate per User
                    .agg(F.countDistinct("song").alias("countSongs"),
                         F.countDistinct("artist").alias("countArtists"),
                         F.collect_list("page").alias("pages"), 
                         (F.sum(F.col("length")) / 3600).alias("totalLengthInHours"),
                         F.first("latestLevel").alias("latestLevel"),
                         F.first("usagePeriod").alias("usagePeriod"),
                         F.max("churned").alias("label")
                        )
                    .withColumn("label", F.when(F.col("label") == True, 1.0).otherwise(0.0))
                    .cache()
                   )
```

```python
cols = aggLogDataByUser.columns
cols.remove("pages")
aggLogDataByUser = (aggLogDataByUser
                    .withColumn("page", F.explode("pages"))
                    .groupBy(cols)
                    .pivot("page")
                    .count()
                    .fillna(0)
                   )
```

```python
def create_lits_list_from_list(lst):
    """ Creates an array of literals from list. """
    return [F.lit(val) for val in lst]


quantileDict = {}

cols = [
    'countSongs',
    'countArtists',
    'totalLengthInHours',
    'usagePeriod',
    'Add Friend',
    'Add to Playlist',
    'Help',
    'NextSong',
    'Roll Advert',
    'Settings',
    'Thumbs Down'
]

for col in cols:
    quantileDict[col] = create_lits_list_from_list(aggLogDataByUser.approxQuantile(col, [0.05, 0.95], 0))
    
```

```python
aggLogDataByUserWoOutlier = aggLogDataByUser
```

```python
# Remove outliers and log scale 
for col in quantileDict.keys():
    aggLogDataByUserWoOutlier = (aggLogDataByUserWoOutlier
                                 .withColumn(col, F.when(F.col(col) < quantileDict[col][0], quantileDict[col][0])
                                                    .when(F.col(col) > quantileDict[col][1], quantileDict[col][1])
                                                    .otherwise(F.col(col))).alias(col)
                                 .fillna(0, subset=[col])
                                )
```

```python
cols = ["totalLengthInHours", 
        "usagePeriod",
        'Add Friend',
        'Add to Playlist',
        'Help',
        'NextSong',
        'Roll Advert',
        'Settings',
        'Thumbs Down']

for col in cols:
    aggLogDataByUserWoOutlier = aggLogDataByUserWoOutlier.withColumn(col, F.log(col)).fillna(0.0, subset=[col])
```

### Feature extraction and Assembly

#### StringIndexer & OneHotEncoder: `latestLevel`

```python
si = StringIndexer(inputCol="latestLevel", outputCol="latestLevel_si")
ohe = OneHotEncoder(inputCol="latestLevel_si", outputCol="latestLevel_ohe")
```

#### VectorAssembler: Assemble all feature columns

```python
features = ["latestLevel_ohe", "countSongs", "countArtists", "totalLengthInHours", "usagePeriod", 
            'Add Friend', 'Add to Playlist', 'Help', 'NextSong', 'Roll Advert', 'Settings',
            'Thumbs Down']

va = VectorAssembler(inputCols=features, outputCol="features")
```

#### Standardize features

```python
ss = StandardScaler(withMean=True, withStd=True, inputCol="features", outputCol="scaledFeatures")
```

## 4 – Data Modeling
In step 4 of the CRISP-DM a model is created and fit to the data, as a first step to answer the 3rd question: Can we actually predict churn from the data? (final answer will follow in Section `5`)

### Question 3: Can we predict churn from those features? 

#### Split data in training and test set

```python
trainValSet, testSet = aggLogDataByUserWoOutlier.randomSplit([0.8, 0.2], seed=42)
```

#### Create the model

```python
lr = LogisticRegression(featuresCol="scaledFeatures", labelCol="label")
```

#### Create Pipelinie

```python
feature_preprocessing = [si, ohe, va, ss]
lrPipeline = feature_preprocessing + [lr]
```

#### Train Model

```python
lrModel = lrPipeline.fit(trainValSet)
```

```python
print(f"f1-score on training set: {evaluate_model(lrModel, trainValSet, 'fMeasureByLabel')}")
print(f"Recall on training set: {evaluate_model(lrModel, trainValSet, 'recallByLabel')}")
print(f"Precision on training set: {evaluate_model(lrModel, trainValSet, 'precisionByLabel')}")
```

```python
lrTrainSetPred = lrModel.transform(trainValSet)
```

```python
confusionMatrixTrainSet = lrTrainSetPred.select("label", "prediction").groupBy("label").pivot("prediction").count()
confusionMatrixTrainSet.show()
```

From the prediction results on the training set, it can be concluded that the model did indeed learn a relation between the input features and the churn label. Of course, this does not mean that the model is actually useful in production, therefore the model will be evaluated on the test set in the next section. 

## 5 – Evaluate the Results
In the final step of the CRISP-DM, the results from the model created in `Step 4` are evaluated. Here, the f1-score was chosen as an evaluation metric due to the imbalance in the data set (fewer users who churned than those who did not). The f1-score is the harmonic mean of precision and recall [Source], therefore also results for those two are provided. 


### Test Model on Test Set

```python
def evaluate_model(model, testSet, metric="fMeasureByLabel"):
    """
    Parameters
    ----------
    model : pyspark.ml model object 
        Specifies a fit model to be evaluated.
    testSet : dataframe
        Dataframe for which prediction are evaluated.
    metric : {"fMeasureByLabel", "f1", "recallByLabel", "precisionByLabel"}, default="fMeasureByLabel"
        Specifies the metric to calculate. Takes all metrics available in MulticlassClassificationEvaluator
    
    Returns
    -------
    The result for the specified metric on the test set.
    
    """
    result = model.transform(testSet)
    
    evaluator = MulticlassClassificationEvaluator(metricName=metric, metricLabel=1.0)
    
    return evaluator.evaluate(result)
```

####  Result

```python
print(f"f1-score: {evaluate_model(lrModel, testSet, 'fMeasureByLabel')}")
print(f"Recall: {evaluate_model(lrModel, testSet, 'recallByLabel')}")
print(f"Precision: {evaluate_model(lrModel, testSet, 'precisionByLabel')}")
```
